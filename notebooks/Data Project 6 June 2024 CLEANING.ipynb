```python
import pandas as pd
data = pd.read_csv('long_format_annual_surface_temp.csv')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data.fillna('unknown', inplace=True)  # Replace NaN with 'unknown'
```

```python
data['Country'] = data['Country'].str.strip()
data['ISO2'] = data['ISO2'].str.strip()
data['Year'] = data['Year'].str.strip()
```

```python
data.rename(columns={'ISO2': 'CountryID'}, inplace=True)
```

```python
# Strip the "F" from the 'Year' column and convert to numeric
data['Year'] = data['Year'].str.lstrip('F').astype(int)
```

```python
data.head()
data.info()
data.describe()
```

```python
data.to_csv('long_format_annual_surface_temp CLEANED.csv', index=False)
```
